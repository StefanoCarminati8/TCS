```python
import numpy as np
import matplotlib.pyplot as plt

# definisco i valori x
x = np.linspace(0, 5, 100)

# definisco i valori y usando la funzione
y = np.exp(-x) * np.cos(2 * np.pi * x)

# creo il grafico
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, linewidth=2, color='blue')

# modifico il grafico
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot della funzione exp(-x) * cos(2*pi*x)')

# mostro il grafico
plt.show()

# salvo il grafico su un file png
plt.savefig('output10.png')
```
