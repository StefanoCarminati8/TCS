```python
import numpy as np
import matplotlib.pyplot as plt

# Carico i dati
data = np.loadtxt("data.dat")

# estraggo i valori di x e y
x = data[:,0]
y = data[:,1]

# creo il grafico 
fig, ax = plt.subplots()
ax.scatter(x, y, color='red')

# modifico il grafico
ax.set_title('Particelle cariche')
ax.set_xlabel('x')
ax.set_ylabel('y')

# salvo il grafico su un file png
plt.savefig('output11.png')

# stampo il grafico
plt.show()
```
