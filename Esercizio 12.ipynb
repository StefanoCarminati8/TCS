```python
import numpy as np
import matplotlib.pyplot as plt

# definisco la funzione
def f(x):
    return -np.sin(x*x)/x + 0.01 * x*x

# genero un array di x
x = np.linspace(-3, 3, 100)

# valuto la funzione in ciascun x
y = f(x)

# scrivo i valori di x e y in un file di output
with open('output.dat', 'w') as file:
    for i in range(len(x)):
        file.write(str(x[i]) + ' ' + str(y[i]) + '\n')

# metto i punti in un grafico
plt.scatter(x, y, color='red')

# aggiungo linea tra i punti
plt.plot(x, y, color='blue')

# sistemo asse delle x
plt.xlim([-3, 3])

# modifico il grafico
plt.title('plot della funzione')
plt.xlabel('x')
plt.ylabel('y')

# aggiungo legenda con equazione della funzione
plt.legend(['$y=-\\frac{sin(x^2)}{x}+0.01x^2$'])

# salvo il grafico in un file di output
plt.savefig('output12.png')

# mostro il grafico
plt.show()
```
